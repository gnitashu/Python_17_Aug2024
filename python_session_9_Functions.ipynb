# Global

- Global keyword is used to take the local variable value outside the function


```python
value = 100
def greet():
    global value
    value=200
    print('hello')

greet()
print('value:',value)    
```

```python
n11=10
def addition():
    global n33,n22
    n22=200
    n33=n11+n22
    print(n11,n22,n33)

addition()
```

```python
n33
```

 - outside the function is called Global variables

 - Inside the function is Local variables

 - Local variable can not use outside the function

 - If you want to use outside the function use **global** keyword
