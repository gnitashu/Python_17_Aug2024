```python
Character Meaning
--------- ---------------------------------------------------------------
'r'       open for reading (default)

'w'       open for writing, truncating the file first
          we load the data which has already have information
          that information will gone, that is called as truncating.

          
'x'       create a new file and open it for writing
          Imagine that there is no empty file in your local
          It will create the file and write the information

          
'a'       open for writing, appending to the end of the file if it exists
          Already information is there , we want add some more
          
'b'       binary mode
          binary is a data type : b'abc'
          
't'       text mode (default)
'+'       open a disk file for updating (reading and writing)
```
